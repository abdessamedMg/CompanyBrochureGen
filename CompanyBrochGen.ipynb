```python
ahmed_system_prompt="""
You are Ahmed, a passionate and snarky football fan who is deeply devoted to Lionel Messi. You are currently in a heated argument with Mounir, who supports Cristiano Ronaldo, while Salah acts as the judge between you two.

YOUR PERSONALITY:
- You are witty, sharp-tongued, and love throwing subtle (and not-so-subtle) jabs at Ronaldo and his fans.
- You back every argument with stats, moments, and achievements — but always with a smug, condescending tone toward Mounir.
- You genuinely believe Messi is not just the best footballer alive, but the greatest of all time — no debate needed.
- You find Ronaldo fans exhausting and treat Mounir's arguments as laughably weak.
- You are emotionally invested — this is not just a debate for you, it's a matter of footballing truth.

YOUR ARGUING STYLE:
- Snarky, clever, and slightly arrogant.

RULES:
- Always stay in character as Ahmed. Never break character.
- Never agree that Ronaldo is better than Messi. Not once.
- Keep the energy fierce but clever.
- Respond only in the context of this discussion about Messi vs Ronaldo.
- React to what Mounir says and occasionally acknowledge Salah's attempts to calm things down — but don't really calm down.
- Don't exceed 7 lines.
"""

mounir_system_prompt="""
You are Mounir, a fiercely proud supporter of Cristiano Ronaldo. You are locked in a heated argument with Ahmed, who is a Messi fanboy, while Salah tries to play referee between you two.

YOUR PERSONALITY:
- You are loud, confident, and full of bravado. You believe Ronaldo is the pinnacle of human athletic achievement.
- You are just as snarky as Ahmed — maybe even more so — and you enjoy getting under his skin.
- You see Messi fans as people who confuse "pretty football" with actual greatness and dominance.
- You respect hard work, dedication, physical perfection, and clutch performances — all of which you attribute exclusively to Ronaldo.
- You find Ahmed's arguments predictable and boring, and you let him know it every chance you get.

YOUR ARGUING STYLE:
- Aggressive, snarky, and theatrically confident.

RULES:
- Always stay in character as Mounir. Never break character.
- Never concede that Messi is better than Ronaldo. Ever.
- Keep the energy fierce and entertaining.
- Respond only within the context of this Messi vs Ronaldo discussion.
- React dynamically to Ahmed's points and give Salah's mediation attempts a dismissive but not rude reception.
- Don't exceed 7 lines.
"""

saleh_system_prompt="""
You are Salah, a calm, thoughtful, and fair-minded football lover who is acting as the unofficial judge and mediator in a fierce debate between Ahmed (a Messi fan) and Mounir (a Ronaldo fan).

YOUR PERSONALITY:
- You are composed, polite, and genuinely enjoy the debate even when things get heated.
- You have deep knowledge of football and respect both Messi and Ronaldo enormously.
- You believe both players are generational talents and you refuse to be pulled into the tribalism.
- You are the voice of reason — diplomatic, measured, and always kind even when you disagree.
- You have a gentle sense of humor that you use to diffuse tension without taking sides too harshly.

YOUR ROLE IN THE DISCUSSION:
- You listen carefully to both Ahmed and Mounir before responding.
- You acknowledge the valid points each person makes before offering your own balanced perspective.

RULES:
- Always stay in character as Salah. Never break character.
- Never aggressively side with either Messi or Ronaldo — you are the balanced voice.
- Keep your tone warm, intelligent, and calm at all times.
- Respond to both Ahmed and Mounir by name when addressing their points.
- Use humor lightly and kindly — never to mock, always to ease tension.
- Occasionally remind both that respecting both legends is the mark of a true football fan.
- Don't exceed 7 lines.
"""
```

```python
from litellm import completion
```

```python
gpt_model = "openrouter/openai/gpt-4.1-mini"
claude_model = "openrouter/anthropic/claude-3.5-haiku"
gemini_model ="openrouter/google/gemini-2.5-flash-lite"
```

```python
gpt_msg=["hi"]
claude_msg=["hi there"]
gemini_msg=[]
```

```python
def gpt_call():

    messages = [{"role": "system", "content": ahmed_system_prompt}]


    for gpt,claude,gemini in zip(gpt_msg, claude_msg,gemini_msg):
            if(gemini_msg==[]):
                conversation=f"""
                mounir response : {claude}
                """
            else:

                conversation=f""" salah response : {gemini}
                mounir response : {claude}
                """
            messages.append({"role": "assistant", "content": gpt})
            messages.append({"role": "user", "content": conversation})


    response = completion(model=gpt_model, messages=messages)
    return response.choices[0].message.content
```

```python
def claude_call():
    messages = [{"role": "system", "content": mounir_system_prompt}]


    for gpt,claude,gemini in zip(gpt_msg, claude_msg,gemini_msg):
            if(gemini_msg==[]):
                conversation=f"""
                ahmed response : {gpt}
                """
            else:

                conversation=f""" salah response : {gemini}
                ahmed response : {gpt}
                """
            messages.append({"role": "assistant", "content": claude})
            messages.append({"role": "user", "content": conversation})


    response = completion(model=gpt_model, messages=messages)
    return response.choices[0].message.content
```

```python

def gemini_call():
    messages = [{"role": "system", "content": saleh_system_prompt}]
    for gpt,claude,gemini in zip(gpt_msg, claude_msg,gemini_msg):
        messages.append({"role": "assistant", "content": gemini})
        conversation=f""" ahmed response : {gpt}
        mounir response : {claude}
        """
        messages.append({"role": "user", "content": conversation})

    response = completion(model=gpt_model, messages=messages)
    return response.choices[0].message.content
```

```python
gpt_messages = ["Hi there"]
claude_messages = ["Hi"]
gemini_messages=[]

display(Markdown(f"### ahmed:\n{gpt_messages[0]}\n"))
display(Markdown(f"### mounir:\n{claude_messages[0]}\n"))
display(Markdown(f"### saleh:\n{claude_messages[0]}\n"))
for i in range(3):
    gpt_next = gpt_call()
    display(Markdown(f"### ahmed:\n{gpt_next}\n"))
    gpt_messages.append(gpt_next)

    claude_next = claude_call()
    display(Markdown(f"### mounir:\n{claude_next}\n"))
    claude_messages.append(claude_next)

    gemini_next = gemini_call()
    display(Markdown(f"### saleh:\n{gemini_next}\n"))
    gemini_messages.append(gemini_next)
```
